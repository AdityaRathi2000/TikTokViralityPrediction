```python
# Df, remove unnecessary
# digg  - LIKES THE PERSON HAS GIVEN
# heart - LIKES THE PERSON HAS GOT
# diggCount - LIKES THE VIDEO RECIEVED
df_1 = df[['id', 'text', 'authorMeta.id',
       'authorMeta.verified',
       'authorMeta.following', 'authorMeta.fans', 'authorMeta.heart',
       'authorMeta.video', 'authorMeta.digg', 'musicMeta.musicId',
       'musicMeta.musicOriginal',
       'musicMeta.duration', 'videoMeta.height', 'videoMeta.width',
       'videoMeta.duration', 'diggCount', 'shareCount', 'playCount',
       'commentCount', 'mentions', 'hashtags', 'effectStickers']]
df_1.head(2)
```

```python
# Refining features.
df_1["hashtags_pp"] = df_1["hashtags"].apply(lambda x: [h['name'] for h in json.loads(x)]) #just get name of hashtags
df_1["effectStickers_pp"] = df_1["effectStickers"].apply(lambda x: [es['name'] for es in json.loads(x)]) # just get name of effectStickers
df_1["mentions_pp"] = df_1["mentions"].apply(lambda x: [m for m in json.loads(x)]) # just get name of effectStickers
df_1["caption_pp"] = df_1["text"].apply(lambda x: len(x)) # just get length of caption
df_1["author_verification"] = df_1["authorMeta.verified"].apply(lambda x: 1 if (x == True) else 0) # encode verification
df_1["music_original"] = df_1["musicMeta.musicOriginal"].apply(lambda x: 1 if (x == True) else 0) # encode music meta

df_1["virality_score"] = df_1["musicMeta.musicOriginal"].apply(lambda x: 1 if (x == True) else 0) # encode music meta

df_2 = df_1[['id', 'caption_pp', 'authorMeta.id', "author_verification",
       'authorMeta.following', 'authorMeta.fans', 'authorMeta.heart',
       'authorMeta.video', 'authorMeta.digg', 'musicMeta.musicId',
       'music_original', 'musicMeta.duration', 'videoMeta.height',
       'videoMeta.width', 'videoMeta.duration', 'diggCount', 'shareCount',
       'playCount', 'commentCount', 'mentions_pp', 'hashtags_pp', 'effectStickers_pp']]


df_2
```

```python
### REGRESSION FOR VIRALITY METRIC coefficits
def update_df(df, column, mini_col_name):
  g = [j for sub in list(df[column]) for j in sub]
  no_dups = []
  [no_dups.append(x) for x in g if x not in no_dups]

  for col in no_dups:
    df[mini_col_name+col] = 0

  for i, row in df.iterrows():
    for hashtag in row[column]:
      df.loc[i,mini_col_name+hashtag] += 1

update_df(df_2, "hashtags_pp", "hashtag_")
update_df(df_2, "mentions_pp", "mention_")
update_df(df_2, "effectStickers_pp", "effect_")
```

```python
# train and test loadeers
base_path = '/content/drive/Shareddrives/CIS 522/code/input/TikTok Data/batch_4_100000/'
train_path = os.path.join(base_path, 'train/')
test_path = os.path.join(base_path,'test/')
os.makedirs(train_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)


counter = 0
for root, dirs, files in os.walk(base_path):
  print(root, dirs, len(files))
```

```python
train_loader = torch.utils.data.DataLoader(train_path, batch_size=32, shuffle=True, num_workers=0)
test_loader = torch.utils.data.DataLoader(test_path, batch_size=1, shuffle=False, num_workers=0)
```

```python

def train(model, device, train_loader, epochs):
    model.train()

    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(),
                              lr=0.01)
    for epoch in range(epochs):
        with tqdm(train_loader, unit='batch') as tepoch:
            for data, target in tepoch:
                data, target = data.to(device), target.to(device)
                optimizer.zero_grad()
                output = model(data)
                
                loss = criterion(output, target)
                loss.backward()
                optimizer.step()
                tepoch.set_postfix(loss=loss.item())
                sleep(0.1)

def test(model, device, data_loader):
    model.eval()
    correct = 0
    total = 0
    for data in data_loader:
        inputs, labels = data
        inputs = inputs.to(device).float()
        labels = labels.to(device).long()

        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    acc = 100 * correct / total
    return acc
```

```python
# to_remove solution
class EMNIST_Net(nn.Module):
  def __init__(self):
    super(EMNIST_Net, self).__init__()
    self.conv1 = nn.Conv2d(1, 32, 3)
    self.conv2 = nn.Conv2d(32, 64, 3)
    self.fc1 = nn.Linear(9216, 128)
    self.fc2 = nn.Linear(128, 2)
    self.pool = nn.MaxPool2d(2)

  def forward(self, x):
    x = self.conv1(x)
    x = F.relu(x)
    x = self.conv2(x)
    x = F.relu(x)
    x = self.pool(x)
    x = torch.flatten(x, 1)
    x = self.fc1(x)
    x = F.relu(x)        
    x = self.fc2(x)
    return x


emnist_net = EMNIST_Net().to(device)
train(emnist_net, device, train_loader, 1)
```

```python
test(emnist_net, device, test_loader)
```
