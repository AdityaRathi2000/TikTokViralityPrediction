```python
# Df, remove unnecessary
# digg  - LIKES THE PERSON HAS GIVEN
# heart - LIKES THE PERSON HAS GOT
# diggCount - LIKES THE VIDEO RECIEVED
df_1 = df[['id', 'text', 'authorMeta.id',
       'authorMeta.verified',
       'authorMeta.following', 'authorMeta.fans', 'authorMeta.heart',
       'authorMeta.video', 'authorMeta.digg', 'musicMeta.musicId',
       'musicMeta.musicOriginal',
       'musicMeta.duration', 'videoMeta.height', 'videoMeta.width',
       'videoMeta.duration', 'diggCount', 'shareCount', 'playCount',
       'commentCount', 'mentions', 'hashtags', 'effectStickers']]
df_1.head(2)
```

```python
# Refining features.
df_1["hashtags_pp"] = df_1["hashtags"].apply(lambda x: [h['name'] for h in json.loads(x)]) #just get name of hashtags
df_1["effectStickers_pp"] = df_1["effectStickers"].apply(lambda x: [es['name'] for es in json.loads(x)]) # just get name of effectStickers
df_1["mentions_pp"] = df_1["mentions"].apply(lambda x: [m for m in json.loads(x)]) # just get name of effectStickers
df_1["caption_pp"] = df_1["text"].apply(lambda x: len(x)) # just get length of caption
df_1["author_verification"] = df_1["authorMeta.verified"].apply(lambda x: 1 if (x == True) else 0) # encode verification
df_1["music_original"] = df_1["musicMeta.musicOriginal"].apply(lambda x: 1 if (x == True) else 0) # encode music meta

df_1["virality_score"] = df_1["musicMeta.musicOriginal"].apply(lambda x: 1 if (x == True) else 0) # encode music meta

df_2 = df_1[['id', 'text', 'caption_pp', 'authorMeta.id', "author_verification",
       'authorMeta.following', 'authorMeta.fans', 'authorMeta.heart',
       'authorMeta.video', 'authorMeta.digg', 'musicMeta.musicId',
       'music_original', 'musicMeta.duration', 'videoMeta.height',
       'videoMeta.width', 'videoMeta.duration', 'diggCount', 'shareCount',
       'playCount', 'commentCount', 'mentions_pp', 'hashtags_pp', 'effectStickers_pp']]


df_2
```

```python
### REGRESSION FOR VIRALITY METRIC coefficits
def update_df(df, column, mini_col_name):
  g = [j for sub in list(df[column]) for j in sub]
  no_dups = []
  [no_dups.append(x) for x in g if x not in no_dups]

  for col in no_dups:
    df[mini_col_name+col] = 0

  for i, row in df.iterrows():
    for hashtag in row[column]:
      df.loc[i,mini_col_name+hashtag] += 1

update_df(df_2, "hashtags_pp", "hashtag_")
update_df(df_2, "mentions_pp", "mention_")
update_df(df_2, "effectStickers_pp", "effect_")
```

```python
df_3 = df_2.drop(["authorMeta.id", "musicMeta.musicId", "mentions_pp", "hashtags_pp", "effectStickers_pp"], axis=1) 
df_3["views/fans"] = df_3["playCount"] / df_3["authorMeta.fans"]

GET_10_PERC_VIRAL = 40 # Multiple 
df_3["label"] = df_3["views/fans"].apply(lambda x: 1 if (x>GET_10_PERC_VIRAL) else 0)
print("1: {}, 0: {}".format(round(len(df_3[df_3["label"] == 1])/len(df_3) * 100, 2), round(len(df_3[df_3["label"] == 0])/len(df_3) * 100, 2)))
df_4 = df_3.drop(["views/fans"], axis=1) 

df_4
```
